# Data visualization

```python
# @title ####Group by driver_gender and driver age
df2 = df[['driver_gender', 'driver_age']]

df3 = df2.groupby('driver_gender')['driver_age'].mean().astype(int).round(0).reset_index()
print(df3)
```

```python
#  @title bar chart that shows average relation between driver_gender and driver_age
# Create a bar chart
plt.figure(figsize=(8, 5))
bars = plt.bar(df3['driver_gender'], df3['driver_age'], color=['#1f77b4', '#ff7f0e'],width=0.4)

# Add data labels
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, int(yval), ha='center', va='bottom')

# Add titles and labels
plt.title('Average Driver Age by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Age')
plt.ylim(0, max(df3['driver_age']) + 5)  # Adjust y-axis limit for better label placement
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Show the plot
plt.show()

```

```python
# @title ##group by Count of Arrest Records by Driver classification

# Group by driver_race and count number of is_arrested for each race
h1=df[['driver_race', 'is_arrested']]
h2 = h1.groupby('driver_race')['is_arrested'].count().reset_index()

# Print the resulting DataFrame
print(h2)

```

# Time Analysis

```python
#  @title how the number of stops changes over time on yearly basis

# Convert 'stop_date' to datetime format
df['stop_date'] = pd.to_datetime(df['stop_date'])

# Resample the data to count the number of stops per year
df_resampled = df.resample('y', on='stop_date').size()

# Plot the resampled data
plt.figure(figsize=(12, 6))
plt.plot(df_resampled.index, df_resampled.values, marker='X', color='darkorange', linestyle='-.', linewidth=2, markersize=6)

# Add data labels
for i, value in enumerate(df_resampled.values):
    plt.text(df_resampled.index[i], value + 0.1, int(value), ha='center', va='bottom')

# Add titles and labels
label_font = {'family': 'serif','color': 'darkred','size': 15}
title_font = {'family': 'serif','color': 'darkblue','size': 18}
plt.title('How the Number of Stops Changes Over Time on a Quarterly Basis',fontdict=title_font)
plt.xlabel('Date',fontdict=label_font)
plt.ylabel('Number of Stops',fontdict=label_font)
plt.xticks(rotation=45)
plt.grid(True) # grid lines
plt.tight_layout() # to make the layout fit the graph


plt.show()
```

**This plot helps me to  visualize periods of higher and lower activity, allowing me to understand insights into seasonal variations or trends in stop frequencies.**
